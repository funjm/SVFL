```python
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
```

```python
class StudentMLP(nn.Module):
    def __init__(self, size):
        super(StudentMLP, self).__init__()
        self.fc1 = nn.Linear(size, size)

    def forward(self, x):
        x = self.fc1(x)
        return x

#     权重初始化为全1,bias为0
    def initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.ones_(m.weight.data)
                m.bias.data.zero_()
                print("\nafter init:  ",m) 
                print('weight is: ', m.weight)
                print(m.bias)
```

```python
size = 3
net_list = list()
net_optimizer_set = list()
for i in range(size):
    net_list.append(StudentMLP(2))
    net_optimizer_set.append(optim.Adam(net_list[i].parameters(), lr=1, betas=(0.9, 0.999)))
```

```python
net = StudentMLP(2)
label = torch.tensor([0., 1], requires_grad=True)
data = torch.tensor([0., 1], requires_grad=True)
print(label)
outputs = net(label)
criterion = nn.MSELoss()
a = criterion(label, outputs)
a
```

```python
print("the net's grad----------", net_list[0].fc1.weight.grad)
net_optimizer_set[0].zero_grad()
print("the net's grad----------", net_list[0].fc1.weight.grad)
```

```python
test_optimizer = optim.Adam([data], lr=1, betas=(0.9, 0.999))
test_scheduler = torch.optim.lr_scheduler.StepLR(test_optimizer, step_size=10, gamma=0.1)
```

```python
for converge in range(100):
    tmp_loss = torch.zeros((size), dtype=torch.float32)
    
    for i in range(size):
        print("the net's grad----------", net_list[i].fc1.weight.grad)
        outputs = net_list[i](data)
        tmp_loss[i] = (criterion(label, outputs))
        print("single loss", tmp_loss[i])
    test_loss = torch.sum(tmp_loss)
    print("test loss: ", test_loss)
    test_loss.backward()
    
    test_optimizer.step()
    print("data's grad===========", data.grad)
    test_optimizer.zero_grad()
    test_scheduler.step()
```
